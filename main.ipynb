```python
import pandas as pd
```

```python
data = pd.read_excel('OutputData.xlsx')
```

```python
data
```

```python
agg_sale = data.groupby(['ProductID', "Name"]).agg({"OrderQty":"sum", "TotalSale":"sum"}).copy()
```

```python
agg_sale.sort_values(by=["TotalSale"], ascending=False, inplace=True)
```

```python
agg_sale
```

```python
agg_sale["CumulativeSum"] = agg_sale["TotalSale"].cumsum()
agg_sale["PrecentageOfTotal"] = 100 * agg_sale["CumulativeSum"] /agg_sale["TotalSale"].sum()
```

```python
agg_sale
```

### ABC Analysis

**From 0 to 20 : 'A'**

**From 20 to 50 : 'B'**

**From 50 to 100 : 'C'**


```python
def ranking(row):

    if row["PrecentageOfTotal"] <= 20:
        return "A"
    
    elif row["PrecentageOfTotal"] > 20 and row["PrecentageOfTotal"] <= 50:
        return "B"
    
    else:
        return "C"
```

```python
agg_sale["ABCRanking"] = agg_sale.apply(ranking, axis=1)
```
